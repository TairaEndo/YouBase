```python
# csvのインポート
df = pd.read_csv('../csv/correction_ALLpitch2018_NPB.csv', encoding="cp932")
dfdic = pd.read_csv('../csv/ID2019.csv',sep=',', encoding="cp932")
```

```python
# 1軍抽出
df = df[df['Level'] == 'NPB']
# 球種名変換辞書
dic1 = {'Fastball': '4シーム', 'Slider': 'スライダー', 'Curveball': 'カーブ', 'ChangeUp': 'チェンジアップ', 'Splitter': 'スプリッター', 'Cutter':'カットボール', 'Other':'不明', 'Sinker':'2シーム', 'Undefined':'不明', 'Knuckleball':'ナックル'}
# 辞書に当てはまるものを置換
df['TaggedPitchType'] = df['TaggedPitchType'].replace(dic1)

"""# 単位を整理
df['PlateLocHeight2'] = df['PlateLocHeight'] * 100
df['PlateLocSide2'] = df['PlateLocSide'] * 100
df['ay0'] = df['ay0'] * 0.03048
df['RelHeight2'] = df['RelHeight'] *100"""

# 不明球を省く
df.drop(df[df['TaggedPitchType'].isin(['不明','ナックル'])].index,inplace=True)
# nullを省く
df.dropna(subset=['TaggedPitchType', 'PlateLocHeight', 'PlateLocSide'],inplace=False)

# 描画の際のラベル
colormap = {'4シーム': '#fc9f9f', 'スライダー': '#6f97fa', 'カーブ': '#a1d780', 'チェンジアップ': '#f7e427', 'スプリッター': '#c280e2', 'カットボール':'#ffb748', '不明':'#ffffff', '2シーム':'#5cd1e4', 'ナックル':'#999999'}
e_colormap = {'4シーム': '#fb3a3a', 'スライダー': '#3a71fb', 'カーブ': '#55c114', 'チェンジアップ': '#e6951d', 'スプリッター': '#a04dca', 'カットボール':'#e27908', '不明':'#999999', '2シーム':'#38a3cc', 'ナックル':'#666666'}
marker_dic = {'4シーム': 'o', 'スライダー': '<', 'カーブ': 'p', 'チェンジアップ': 'D', 'スプリッター': 's', 'カットボール':'v', '不明':'*', '2シーム':'>', 'ナックル':'*'}

df['color']=[colormap[x] for x in df['TaggedPitchType']]
df['e_color']=[e_colormap[x] for x in df['TaggedPitchType']]
df['marker']=[marker_dic[x] for x in df['TaggedPitchType']]
```

```python
# IDと日本語名の辞書を作成
players = []
players_dict = {}
players_dict2 = {}
with open('../csv/ID2019.csv', 'r', encoding='cp932') as f: #もしこのセルでエラーが出たら、encoding='cp932'　を消すと良いかもしれません
    reader = csv.DictReader(f)
    for r in reader:
        players.append({int(r['id']): r['name']})
        players_dict[int(r['id'])] = r['name']
        # 本当はこういうのがかっこいい
        players_dict2 = {int(row['id']): row['name'] for row in reader}
        
# 置換
df['Pitcher_J'] = df['PitcherId'].replace(players_dict2)
df['Batter_J'] = df['BatterId'].replace(players_dict2)
```

```python
# 二人のジョンソンを区別
df['Pitcher_J'] = np.where(df['Pitcher'] == 'Johnson, Pierce', 'ジョンソン(神)', df['Pitcher_J'])
df['Pitcher_J'] = np.where(df['Pitcher'] == 'Johnson, Kris', 'ジョンソン(広)', df['Pitcher_J'])
```

```python
#ストレートを抽出
df_4seam = df[df['TaggedPitchType'] == '4シーム']
```

```python
def zone_n_cut(N, df):
    
    """
    ストライクゾーンを　N^2 個のゾーンに分割、
    ボールゾーンを含めると (N+2)^2 個のゾーンに分割し
    それぞれのゾーンに番号と名前を割り当てます
    
    入力  N : ストライクゾーンの分割数
             df : ['PlateLocHeight', 'PlateLocSide']の2つのカラムを含んだトラックマンデータのDF
             
    出力: df_zone_concat : 入力されたDFに['VertCut', 'HorzCut', 'ZoneNum', 'ZoneName']の4つのカラムを追加する
            <カラム詳細>
            VertCut : 縦方向にN分割した時、何番目のゾーンに属しているか。1〜(N+2)の値をとる
            HorzCut : 横方向にN分割した時、何番目のゾーンに属しているか。1〜(N+2)の値をとる
            ZoneNum : 右打者の内角高めから外角低めに向かって順に番号をつける。1〜(N+2)^2をとる。ストライクとボールの区別はつけない。(int型)
            ZoneName : ボールゾーンはB、ストライクゾーンはSを数字の前につけて区別した。
                                 それぞれ右打者の内角高めから外角低めに向かって順に番号が振られる(str型)
            
            
    ※注意 :  'PlateLocHeight', 'PlateLocSide'のどちらかがNaNの行は削除しています。
    """
    
    #ストライクゾーンの縦横閾値(大体で決めました)
    vert_stmin, vert_stmax = 0.5, 1.0
    horz_stmin, horz_stmax = -0.375, 0.375
    
    #ストライクゾーンの縦横長さ
    vert_length = vert_stmax - vert_stmin
    horz_length = horz_stmax - horz_stmin
    
    #長さをN分割し、１つのゾーンの縦横長さを求める
    vert_length_unit = vert_length / N
    horz_length_unit = horz_length / N
    
    #縦横ストライクゾーンの下限から1ゾーン分だけボールゾーンにずらしたものを、ボールゾーンの下限とする
    vert_min = vert_stmin - vert_length_unit
    horz_min = horz_stmin - horz_length_unit
    
    #縦横ストライクゾーンの上限から1ゾーン分だけボールゾーンにずらしたものを、ボールゾーンの上限とする
    vert_max = vert_stmax + vert_length_unit
    horz_max = horz_stmax + horz_length_unit
    
    #'PlateLocHeight', 'PlateLocSide'のいずれかに欠損値が含まれていたら、その行を削除
    df = df.dropna(subset=['PlateLocHeight', 'PlateLocSide'], axis=0, how="any")
    
    #bin分割のため、座標を反転させる
    df_zone = pd.DataFrame()
    df_zone['PlateLocHeight_reverse'] = - df['PlateLocHeight']
    df_zone['PlateLocSide_reverse'] = - df['PlateLocSide']
    
    #縦横それぞれビン分割 → 作ってから気づいたけど、labelを反転させれば済む話だった。。
    df_zone['VertCut'] = pd.cut(df_zone['PlateLocHeight_reverse'], np.linspace(-vert_max, -vert_min, N+3), labels=[i for i in range(1, N+3)])
    df_zone['HorzCut'] = pd.cut(df_zone['PlateLocSide_reverse'], np.linspace(- horz_max, -horz_min, N+3), labels=[i for i in range(1, N+3)])
    
    #'VertCut', 'HorzCut'のいずれかがNaNの行は削除
    df_zone_concat = pd.concat([df, df_zone], axis=1)
    df_zone_concat = df_zone_concat.dropna(subset=['VertCut', 'HorzCut'], axis=0, how="any")
    
    #VertCut, HorzCutのカテゴリ値をint型に変換
    df_zone_concat.VertCut = df_zone_concat.VertCut.astype(int)
    df_zone_concat.HorzCut = df_zone_concat.HorzCut.astype(int)
    
    #ゾーンナンバーをつける(型はintです, 右上から1〜(N+2)^2まで番号を振りました)
    df_zone_concat['ZoneNum'] = (df_zone_concat.VertCut - 1) * (N+2) + df_zone_concat.HorzCut
    
    #ボールゾーンに名前をつける(型はstrです)
    #df_zone_concat['ZoneName'] = "B" + df_zone_concat['ZoneNum'].astype(str)
    
    #ストライクゾーンに名前をつける(型はstrです)
    query_str = '(VertCut not in [1, @N+2]) and (HorzCut not in [1, @N+2])'
    df_subset = df_zone_concat.query(query_str)
    strike_num = df_zone_concat['ZoneNum'] - (2 * df_zone_concat['VertCut'] + 2)
    df_zone_concat.loc[df_subset.index, 'ZoneName'] = "S" + strike_num.astype(str)
    
    #ボールゾーンに名前をつける(型はstrです)
    #高め
    query_str = 'VertCut == 1'
    df_subset = df_zone_concat.query(query_str)
    ball_high_num = df_zone_concat['ZoneNum']
    df_zone_concat.loc[df_subset.index, 'ZoneName'] = "B" + ball_high_num.astype(str)
    
    #右打者の内角側
    side_range = [i for i in range(2, N+2)]
    df_subset = df_zone_concat.query('(VertCut == @side_range) and (HorzCut == 1)')
    ball_in_num = df_zone_concat['ZoneNum'] - (3*(df_zone_concat['VertCut'] - 2))
    df_zone_concat.loc[df_subset.index, 'ZoneName'] = "B" + ball_in_num.astype(str)
    
    #右打者の外角側
    df_subset = df_zone_concat.query('(VertCut == @side_range) and (HorzCut == @N+2)')
    ball_out_num = df_zone_concat['ZoneNum'] - (3*(df_zone_concat['VertCut'] - 1))
    df_zone_concat.loc[df_subset.index, 'ZoneName'] = "B" + ball_out_num.astype(str)
    
    #低め
    df_subset = df_zone_concat.query('VertCut == @N+2')
    ball_low_num = df_zone_concat['ZoneNum'] - N*2
    df_zone_concat.loc[df_subset.index, 'ZoneName'] = "B" + ball_low_num.astype(str)
    
    

    #不必要なカラムを削除
    df_zone_concat = df_zone_concat.drop(['PlateLocHeight_reverse', 'PlateLocSide_reverse'], axis=1)
    
    return df_zone_concat
```

```python
#ストライクゾーンを3x3に分割する
df_4seam_withZone = zone_n_cut(3, df_4seam)
```

```python
#右バッターには、外角低めを投げているか？(菅野で検証)
df_withZone = zone_n_cut(3, df)
sugano_rightbatter = df_withZone.query('Pitcher_J == "菅野 智之" and BatterSide == "Right"')
```

```python
#コースごとの球数を表示
sugano_rightbatter.groupby('ZoneName').count()['PitchNo'].sort_values(ascending=False)
#→ S6とS9、つまり外角真ん中から低めにボールが集まっている
```

```python
#PlateLocSideがどっち向きなのかを、打球の飛んだ方向を使って検証してみる
#ロペスは引っ張り打者だから、Directionがどちらかに偏っているはず
```

```python
#ロペスの打席を取得
chamo = df_4seam_withZone[df_4seam_withZone['Batter_J'] == 'ロペス']
```

```python
#打球の飛んだ方向を可視化
sns.distplot(chamo.Direction.dropna())
# → マイナスに偏っているので、Directionはマイナスがレフト方向・プラスがライト方向だとわかる
```

```python
#投球の横座標と、打球の飛んだ方向の関係を可視化
chamoDirection = chamo[['PlateLocSide', 'Direction']].dropna()
sns.jointplot(chamoDirection['PlateLocSide'], chamoDirection['Direction'])

# → 横座標がプラスの時に引っ張っているので、右打者にとってはマイナスが外角・プラスが内角だとわかる
```
